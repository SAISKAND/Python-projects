```python
audio, sample_rate = sf.read('/content/drive/MyDrive/Data for ML lab/exp1/audio/whistling.wav')
import matplotlib.pyplot as plt
import numpy as np
time = np.arange(0, len(audio)) / sample_rate

plt.plot(time, audio)
plt.title('Sound wave')
plt.xlabel('Time ')
plt.ylabel('Amplitude')
plt.show()
```
